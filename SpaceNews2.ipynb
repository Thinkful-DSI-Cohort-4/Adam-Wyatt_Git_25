```
from time import time
from time import sleep
from random import randint
from IPython.core.display import clear_output
from warnings import warn
from bs4 import BeautifulSoup
import requests


# inputs
url = "https://www.space.com/news"
max_page = 3
# page function
def scrapper(url, max_page):
  articles = {}
  for page in range(max_page):
    p = page + 1
    articles.update(page_scrapper(url,p))
  return(articles)
# scrapper function
def page_scrapper(url, page):
  url2 = url +'/' + str(page)
  articles = {}
  response = requests.get(url2)
  if(response.ok):
    data = response.text
    soup = BeautifulSoup(data, 'html.parser')
    soup2 = soup.find_all(class_='content')
    for x in soup2:
      headline = x.find(class_='article-name').getText()
      author = x.find(class_='by-author').getText().strip('\n').strip('By').strip('\n\n')[:-1]
      synopsis = x.find(class_='synopsis').getText().strip('\n')
      date = x.find(class_='published-date relative-date').get_attribute_list('data-published-date')[0]
      articles[headline] = [author, synopsis, date]
  sleep(1)
  return(articles)
scrapper(url, max_page)

```
