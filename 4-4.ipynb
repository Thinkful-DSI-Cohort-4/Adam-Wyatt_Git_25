```
import requests
from bs4 import BeautifulSoup

def get_news():
  url= 'https://www.space.com/news'
  response= requests.get(url)
  data= response.text
  soup= BeautifulSoup(data,'html.parser')
  news= soup.find_all('div',class_='content')
  for elem in news:
    headline= elem.find('h3',class_='article-name').get_text().strip()
    author= elem.find('span',class_='by-author').get_text().strip().replace('By','')
    date= elem.find('time',class_='published-date relative-date')
    date= date['datetime']
    synopsis= elem.find('p',class_='synopsis').get_text().strip()
    print(headline)
    print(author)
    print(synopsis)
    print(date)
    print('----------------------------','\n')

get_news()





```
